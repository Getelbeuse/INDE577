# This section presents the implemention and the analysis of the K-means. 


## In the first part:
1. For the K-means clustering, we use sklearn's make_blobs dataset generation tool to create a synthetic dataset suitable for clustering. This dataset contains multiple "clusters" or groups, which clearly demonstrates the effectiveness of the K-means clustering algorithm.

2. we display the data sample distribution. 

```python
# Import necessary libraries for data handling and visualization
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data using make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Plotting the generated data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, c='blue', marker='o', edgecolor='black')
plt.title('Visualization of Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## In the second part:

1. we setup and train the K-means models

2. we display the clustering results in different colors, where each color represents a cluster. At the same time, we also mark the center points of each cluster with a red 'x'. These center points are the "center" of each cluster found by the algorithm.

### Note: the warining here is not improtant


```python
# Import KMeans from sklearn
from sklearn.cluster import KMeans


# Apply K-means clustering
kmeans = KMeans(n_clusters=4, n_init=10, random_state=0)
kmeans.fit(X)

# Predict the cluster labels
y_kmeans = kmeans.predict(X)

# Centers of the clusters
centers = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', marker='o', edgecolor='black')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.title('Clusters and Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Analysis part:
From the above results, we can see that ther clustering results is quite good. This is beacuse we generate the dataset which is naturely suitable for clustering and the cluster number are matched.

Here, we would like to intentionally create some difficulties for the clustering.

### 1. We create the dataset with 4 clusters while enfore the model to cluster them into 3 clusters.

Based on the results, we can observe that the two neighboring clusters are classified into one cluster.
This result is quite reasonable.

```python
# Import necessary libraries for data handling and visualization
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data using make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.70, random_state=0)

# Plotting the generated data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, c='blue', marker='o', edgecolor='black')
plt.title('Visualization of Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()


# Apply K-means clustering
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0)
kmeans.fit(X)

# Predict the cluster labels
y_kmeans = kmeans.predict(X)

# Centers of the clusters
centers = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', marker='o', edgecolor='black')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.title('Clusters and Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### 2. We create the dataset with 4 clusters while enfore the model to cluster them into 5 clusters.

Based on the results, we can observe that the above  cluster is divided into two parts. 
It can be a feasible clustering result, however, more different results can be obtained by setting the hyperparameters.

This results also demonstrate the weakness of the unsupervised learning, which is not reliable, highliting the importance of high-quality labeled data.

```python
# Import necessary libraries for data handling and visualization
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data using make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.70, random_state=0)

# Plotting the generated data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, c='blue', marker='o', edgecolor='black')
plt.title('Visualization of Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()


# Apply K-means clustering
kmeans = KMeans(n_clusters=5, n_init=10, random_state=0)
kmeans.fit(X)

# Predict the cluster labels
y_kmeans = kmeans.predict(X)

# Centers of the clusters
centers = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', marker='o', edgecolor='black')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.title('Clusters and Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
